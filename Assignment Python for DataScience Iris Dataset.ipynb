```python
# Visualization 1: Pairplot – relationship between features

sns.pairplot(df, hue='Classification')
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
# Visualization 2: Boxplot – distribution of sepal length for each class
sns.boxplot(x='Classification', y='SL', data=df)
plt.title("Sepal Length Distribution by Class")
plt.show()
```

```python
# Visualization 3: Heatmap – correlation between features
sns.heatmap(df.drop(columns='Classification').corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
