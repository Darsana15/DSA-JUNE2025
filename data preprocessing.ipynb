```python
import pandas as pd

# 1️⃣ Select all numeric columns
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns

# 2️⃣ Exclude the target variable
target = 'Sale Price'  # replace with your target column
columns_to_review = numeric_cols.drop(target)
print("Numeric columns to review:\n", columns_to_review)

# 3️⃣ Show basic statistics to review scales
stats = data[columns_to_review].describe().T[['min', 'max', 'mean', 'std']]
print("\nColumn statistics:\n", stats)

# 4️⃣ Identify columns with very different scales or high variance
# Adjust thresholds if needed
columns_need_scaling = stats[(stats['max'] > 100) | (stats['std'] > 50)].index.tolist()
print("\nColumns that may benefit from scaling:\n", columns_need_scaling)

# 5️⃣ Check if any numeric column is actually categorical
for col in columns_to_review:
    unique_vals = data[col].nunique()
    total_rows = len(data)
    if unique_vals / total_rows < 0.05:  # less than 5% unique values
        print(f"Column '{col}' may be categorical with {unique_vals} unique values.")
```
